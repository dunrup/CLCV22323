```python
# this 'decodes' the url, which makes it a readable json file
# with urllib.request.urlopen(uri) as f:
#     data = json.loads(f.read().decode()) 
```

```python
def getjson(uri) :
    uri = f'https://pleiades.stoa.org/places/{uri}/json'
    with urllib.request.urlopen(uri) as f:
        data = json.loads(f.read().decode())
    return data
```

```python
getjson(589918)
```

```python
type(getjson(589918))
```

```python
def coordinates(uri) : 
    return getjson(uri)['features'][0]['geometry']['coordinates']
```

```python
coordinates(589918)
```
